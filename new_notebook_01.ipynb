### Nettoyage des données<a id="nettoyage">

__`Traitement des données sur les immatriculations des véhicules neufs et d'occasions 2010 à 2019`__

```python
#Chargement du jeu de données "immatriculation-voitures-neuves-source-energie-localisation"
xls = pd.ExcelFile('p8_data/immatriculation-voitures-neuves-source-energie-localisation.xlsx')
df2010 = pd.read_excel(xls, '2010 ', skiprows=2)
df2011 = pd.read_excel(xls, '2011', skiprows=2)
df2012 = pd.read_excel(xls, '2012', skiprows=2)
df2013 = pd.read_excel(xls, '2013', skiprows=2)
df2014 = pd.read_excel(xls, '2014', skiprows=2)
df2015 = pd.read_excel(xls, '2015', skiprows=2)
df2016 = pd.read_excel(xls, '2016', skiprows=2)
df2017 = pd.read_excel(xls, '2017', skiprows=2)
df2018 = pd.read_excel(xls, '2018', skiprows=2)
df2019 = pd.read_excel(xls, '2019', skiprows=2)
```

```python
#Visualisation d'un jeu de données
df2011.head()
```

```python
#Ajout d'une colonne 'Année' par dataframe
df2010['Année'] = 2010
df2011['Année'] = 2011
df2012['Année'] = 2012
df2013['Année'] = 2013
df2014['Année'] = 2014
df2015['Année'] = 2015
df2016['Année'] = 2016
df2017['Année'] = 2017
df2018['Année'] = 2018
df2019['Année'] = 2019
```

```python
#Concaténation par méthode .concat() des précédents jeux de données
immat_motor_puiss = pd.concat([df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019], 
                               sort=False)
```

```python
#Premier aperçu du jeu de données 
immat_motor_puiss
```

```python
#Suppression des lignes dont les valeurs sont manquantes sur l'objet Series 'Départements'
#Suppression de la variable 'Unnamed: 13' non représentative dans l'échantillon
immat_motor_puiss.dropna(subset=['Départements'], axis=0, inplace=True)
immat_motor_puiss.drop(columns=['Unnamed: 13'], inplace=True)
```

```python
#Renommage des colonnes prenant en compte les puissances de la ligne index 0
immat_motor_puiss.rename(columns={
    'Essence + Superéthanol': 'Essence + Superéthanol < 6CV', 'Unnamed: 3': 'Essence + Superéthanol 6 à 7CV',
    'Unnamed: 4': 'Essence + Superéthanol 8CV et +', 'Electricité-Essence': 'Hybride essence < 6CV',
    'Unnamed: 6': 'Hybride essence 6CV et +', 'Gazole': 'Gazole < 6CV', 'Unnamed: 8': 'Gazole 6CV et +',
    'Essence-GPL': 'Essence-GPL < 6CV', 'Unnamed: 10': 'Essence-GPL 6CV et +', 
    'Gazole-Electricité': 'Hybride gazole', 'Autres et non dét.': 'Autres'
                                  }, inplace=True)
```

```python
#Réindexation du dataframe
immat_motor_puiss.reset_index(drop=True, inplace=True)
```

```python
#Affichage des modalités liées aux départements
immat_motor_puiss['Départements'].unique()
```

Il est nécessaire de supprimer la modalité 'Total' de la variable 'Départements'.

```python
#Boucle itérative de suppression de la modalité 'Total'
for i, row in immat_motor_puiss.iterrows():
    if row['Départements'] == 'Total':
        immat_motor_puiss.drop(i, inplace=True)
        

#Autre méthode
'''immat_motor_puiss.drop(immat_motor_puiss[immat_motor_puiss['Départements'] == 'Total']
                        .index, axis=0, inplace = True) 
'''
```

Rappelons le contexte, le dataframe est constitué de 8 jeux de données traitant des immatriculations des véhicules neufs et d'occassion en France Métropolitaine + Corse de 2010 à 2019. Il peut-être nécessaire d'avoir une vision plus précise sur le positionnement dans le temps et par catégorie de véhicules des observations manquantes.

```python
#Suppression de la variable 'Autres' qui n'est pas exploitable dans l'état et non représentative 
immat_motor_puiss.drop(columns=['Autres'], inplace=True)
```

```python
#Visualisation par restriction des valeurs manquantes de la variable 'Hybride essence 6CV et +'
immat_motor_puiss[(immat_motor_puiss['Hybride essence 6CV et +'].isna())]
```

Notons que suivant l'année tous les types de véhicules n'étaient pas ou peu réprensentés suivant la localité. A ce stade de l'étude, les valeurs manquantes peuvent-être laissées en l'état.

```python
#Création d'une variable 'Essence'
immat_motor = immat_motor_puiss.copy().reset_index()
immat_motor['Essence'] = immat_motor['Essence + Superéthanol < 6CV']\
                         + immat_motor['Essence + Superéthanol 6 à 7CV']\
                         + immat_motor['Essence + Superéthanol 8CV et +']
```

```python
#Création d'une variable 'Gazole'
immat_motor['Gazole'] = immat_motor['Gazole < 6CV'] + immat_motor['Gazole 6CV et +']
```

```python
#Création d'une variable 'Hybride'
immat_motor['Hybride'] = immat_motor['Hybride essence < 6CV']\
                       + immat_motor['Hybride essence 6CV et +']\
                       + immat_motor['Hybride gazole']
```

```python
#Création d'une variable 'GPL'
immat_motor['GPL'] = immat_motor['Essence-GPL < 6CV'] + immat_motor['Essence-GPL 6CV et +']
```

```python
#Projection sur les variables globales
immat_motor = immat_motor[['Régions', 'Départements', 'Année', 'Electricité', 'Essence', 'Gazole', 
                           'Hybride', 'GPL', 'Total']]
```

```python
#Aperçu du jeu de données
immat_motor
```

```python
#Situation des valeurs manquantes par variable
immat_motor.isna().sum()
```

Rappelons que ces 960 observations s'échelonnent sur un timeframe de 2010 à 2019, il est normal d'avoir des NaN suivant le couple motorisation/année, également pouvant s'expliquer par des anomalies de recensement. Rappelons également que l'état des modalités de la variable 'Régions' est le résultat non seulement du remaniement des régions de 2014, mais également de la structuration des données réparties par départements. Dans l'immédiat seule l'approche départementale est pertinente, les modalités liées aux régions peuvent rester sans modification particulière.

```python
#Affichage des observations par agrégation par Nombre d'immatriculation (sans cumul) selon les années
immat_annee = immat_motor.groupby('Année').sum().astype(int)
immat_annee
```

```python
immat_annee.reset_index(inplace=True)
immat_annee['Année'] = pd.to_datetime(immat_annee['Année'], format='%Y')
immat_annee['Année'] = pd.date_range(start='31/12/2010', end='31/12/2019', freq ='Y')
```

```python
#Données des immatriculations par année sans cumul
immat_annee
```

Voyons également en cumulé ce que représentent ces immatriculations…

```python
#Sauvegarde 
immat_annee.to_csv('p8_datatable/immat_annee.csv')
```

```python
#Affichage des observations par agrégation par somme cumulative selon les années
immat_motor_annee = immat_motor.groupby('Année').sum()
immat_motor_annee = immat_motor_annee.cumsum(axis=0).astype(int)
immat_motor_annee
```

```python
immat_motor_annee.reset_index(inplace=True)
immat_motor_annee['Année'] = pd.to_datetime(immat_motor_annee['Année'], format='%Y')
immat_motor_annee['Année'] = pd.date_range(start='31/12/2010', end='31/12/2019', freq ='Y')
```
